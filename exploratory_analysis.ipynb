```python
getpattern_json = r.json()
pattern_route29 = getpattern_json["bustime-response"]["ptr"]
```

```python
stop_num = 7
route = {}
for pattern in pattern_route29:
    stops = {}
    last_stop = "start"
    seq_stop = 1
    for point in pattern["pt"]:
        if point["typ"] == "S":
            stops.update({point["stpid"]:{
                "latitude":point["lat"],
                "longitude":point["lon"],
                "stop_name":point["stpnm"]
            }})
            if last_stop != "start":
                stops[last_stop].update({"next_stop":point["stpid"]})
            last_stop = point["stpid"]
    stops[last_stop].update({"next_stop":"end"})
    route.update({pattern['rtdir'].lower():stops})
print(route)
```
